Mean Median ve Mod ile Python Örnekleri

```python
import numpy as np
```

```python
#Random kullanarak sample bir veri setini kendimiz oluşturalım
#Aylık geliri 5000 tl olan standart sampması 2000tl olan 1000 kişinin random aylık maaşını hesaplama

gelirler=np.random.normal(5000,2000,1000)
gelirler
```

```python
#ortalamayı mean ile hesaplarız
np.mean(gelirler)
```

```python
#bunu çizdireceğiz, bu sebeple plt yi import ediyoruz
import matplotlib.pyplot as plt

plt.hist(gelirler, 100)
plt.show()
```

```python
#ortanca değeri bulmak için median kullanılır

np.median(gelirler)
```

```python
#mode ile en çok tekrar edilen veri döndürülür

#şimdi örnek olarak bir okuldaki çocukların ortalama yaş değerlerini random oluşturalım
#toplam 300 öğrenci olsun ve yaşlar 7 ile 18 arasında

yaslar=np.random.randint(7,18, size=300)

from scipy import stats
stats.mode(yaslar)
```

```python
#en sık değer = 7 ve tekrar sayısı = 36
```

```python
np.median(yaslar)
```

```python
np.mean(yaslar)
```
